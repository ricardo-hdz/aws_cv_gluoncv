```python
image = mx.image.imread('mt_baker.jpg')
```

```python
print(image.shape)
```

```python
plt.imshow(image.asnumpy())
```

```python
image = gcv.data.transforms.presets.imagenet.transform_eval(image)
```

```python
print(image.shape)
```

```python
network = gcv.model_zoo.get_model('ResNet50_v1d', pretrained=True)
```

```python
prediction = network(image)
```

```python
print(prediction.shape)
```

```python
prediction = prediction[0]
```

```python
probability = mx.nd.softmax(prediction)
```

```python
rounded_probability = mx.nd.round(probability*100)/100
print('Skipping 950 values...')
```

```python
print(rounded_probability[950:])
```

```python
mx.nd.softmax(mx.nd.array((-0.5, 0.5)))
```

```python
mx.nd.softmax(mx.nd.array((-1.0, 1.0)))
```

```python
k = 5
topk_indices = mx.nd.topk(probability, k=k)
topk_indices
```

```python
network.classes[980]
```

```python
for i in range(k):
    class_index = topk_indices[i].astype('int').asscalar()
    class_label = network.classes[class_index]
    class_probability = probability[class_index]
    print('#{} {} ({:0.3}%)'.format(i+1, class_label, class_probability.asscalar()*100))
```

### Image Segmentation

```python
image = mx.image.imread('./dog.jpg')
print(image.shape, image.dtype, image.min().asscalar(), image.max().asscalar)
```

```python
plt.imshow(image.asnumpy())
```

```python
from mxnet.gluon.data.vision import transforms
```

```python
transform_fn = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize([.485,.456,.406],[.229,.224,.225])
])
```

```python
image = transform_fn(image)
print(image.shape, image.dtype, image.min().asscalar(), image.max().asscalar)
```

```python
image = image.expand_dims(0)
print(image.shape)
```

```python
network = gcv.model_zoo.get_model('fcn_resnet50_ade', pretrained=True)
```

```python
output = network.predict(image)
print(output.shape)
```

```python
output = output[0]
print(output.shape)
```

```python
px_height, px_width = 300,500
```

```python
px_logit = output[:,px_height, px_width]
```

```python
px_probability = mx.nd.softmax(px_logit)
px_rounded_probability = mx.nd.round(px_probability*100)/100
print(px_rounded_probability)
```

```python
class_index = mx.nd.argmax(px_logit, axis=0)
# finds most likelt class
class_index = class_index[0].astype('int').asscalar()
print(class_index)
```

```python
from gluoncv.data.ade20k.segmentation import ADE20KSegmentation
class_label = ADE20KSegmentation.CLASSES[class_index]
print(class_label)
```

```python
output_proba = mx.nd.softmax(output, axis=0)
output_heatmap = output_proba[127]
plt.imshow(output_heatmap.asnumpy())
```

```python
prediction = mx.nd.argmax(output, 0).asnumpy()
print(prediction.shape)
```

```python
print(prediction)
```
